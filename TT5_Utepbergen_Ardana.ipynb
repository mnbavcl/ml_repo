```python
import numpy as np
from scipy import stats
from sklearn.preprocessing import MinMaxScaler
```

```python
scores = np.array([75, 82, 68, 91, 79, 95, 89, 70, 92])
average_score = scores.mean()
print("Ұпайлар:", scores)
print("Орташа мән:", average_score)
```

```python
A = np.array([1.5, 2.0, 0.5])
B = np.array([3.0, 1.0, 2.5])
C = A + B
print("A =", A)
print("B =", B)
print("C = A + B =", C)
```

```python
v = np.array([2, -1, 5])
scalar = 3
u = scalar * v
print("v =", v)
print("Скаляр:", scalar)
print("u = 3 * v =", u)
```

```python
data_points = np.array([10, 20, 50, 40, 30]).reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 1))
data_normalized = scaler.fit_transform(data_points)
print("Бастапқы мәндер:\n", data_points.ravel())
print("Нормализацияланған мәндер:\n", data_normalized.ravel())
```
