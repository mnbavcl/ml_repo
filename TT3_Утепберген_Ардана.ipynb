```python
file_path = r"C:\Users\ardan\ml\week 3\Salaries.csv"
data_rows = []
with open(file_path, "r", encoding="utf-8") as f:
    lines = f.readlines()
for line in lines[1:11]:   
    data_rows.append(line.strip().split(","))
print("data_rows ұзындығы:", len(data_rows))
print("Бірінші қызметкер:", data_rows[0])
```

```python
first_three_names = (
    data_rows[0][1],
    data_rows[1][1],
    data_rows[2][1]
)
employee_titles = {
    data_rows[0][1]: data_rows[0][2],
    data_rows[1][1]: data_rows[1][2],
    data_rows[2][1]: data_rows[2][2]
}
print("Кортеж:", first_three_names)
print("Сөздік:", employee_titles)
```

```python
base_pays_str = [row[4] for row in data_rows]
print(base_pays_str)
```

```python
job_titles_list = []
for row in data_rows:
    job_titles_list.append(row[2])  
unique_titles = set(job_titles_list)
print("Жұмыс Атақтары:", unique_titles)
print("Жиын пайдалы себебі ол қайталанатын элементтерді жояды")
```
